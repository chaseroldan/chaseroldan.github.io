```python
from sqlalchemy import create_engine
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost/olympics_db')
conn = engine.connect()
```

```python
df = pd.read_sql("SELECT * FROM olympics", conn)
df.head()
```

```python
# Selecting the year 2016 shows only Summer Olympics so season does not need to be specified
```

```python
# Select all USA 2016 gold medalists
USA_Gold_2016 = pd.read_sql("SELECT * FROM olympics WHERE year = '2016' AND team = 'United States' AND medal = 'Gold'", conn)
USA_Gold_2016.head()
```

```python
USA_Gold_2016_Name = USA_Gold_2016['name'].value_counts()
USA_Gold_2016_Name = pd.DataFrame(USA_Gold_2016_Name)
USA_Gold_2016_Name.head(10)
```

```python
# Create bar chart for USA 2016 gold medalists
USA_Gold_2016_Name_Chart = USA_Gold_2016_Name.head(10).plot(kind='bar', figsize=(18,9), rot=10, color='gold')

# Class methods
USA_Gold_2016_Name_Chart.set_ylabel("Medals Won")
USA_Gold_2016_Name_Chart.set_title("USA Gold Medals 2016 - Athlete Name")
plt.savefig('images/USA_Gold_2016_Name.png')
plt.show()
plt.tight_layout()
```

```python
USA_Gold_2016_Sports = USA_Gold_2016['sport'].value_counts()
USA_Gold_2016_Sports = pd.DataFrame(USA_Gold_2016_Sports)
USA_Gold_2016_Sports
```

```python
# Create a bar chart for USA 2016 gold medal sports
USA_Gold_2016_Sports_Chart = USA_Gold_2016_Sports.plot(kind='bar', figsize=(16,7), rot=0, color='gold')

# Class methods
USA_Gold_2016_Sports_Chart.set_ylabel("Medals Won")
USA_Gold_2016_Sports_Chart.set_title("USA Gold Medals 2016 - Sport")
plt.savefig('images/USA_Gold_2016_Sports.png')
plt.show()
plt.tight_layout()
```
